```python
import pandas as pd
import matplotlib.pyplot as plt 

# Load dataset
df = pd.read_csv("../data/filtered_data.csv")
```

```python
from imblearn.over_sampling import ADASYN

# X: features, y : target 
X = df.drop('Churn', axis=1)
y = df['Churn']

# ADASYN
adasyn = ADASYN(random_state=42)
X_adasyn, y_adasyn = adasyn.fit_resample(X, y)

print("Before ADASYN:")
print(y.value_counts())

print("\nAfter ADASYN:")
print(y_adasyn.value_counts())
```

```python
# New dataframe
df_resampled = pd.concat([X_adasyn, y_adasyn], axis=1)

# Saving as CSV
df_resampled.to_csv("../data/resampled_datas/resampled_adasyn.csv", index=False)
```

```python
before_counts = y.value_counts().sort_index()
after_counts = y_adasyn.value_counts().sort_index()

before_colors = ['lightblue' if label == 0 else 'lightcoral' for label in before_counts.index]
after_colors = ['lightblue' if label == 0 else 'lightcoral' for label in after_counts.index]

fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Before
ax[0].bar(before_counts.index.astype(str), before_counts.values, color='lightblue')
ax[0].set_title("Before Resampling")
ax[0].set_xlabel("Churn")
ax[0].set_ylabel("Count")

# After
ax[1].bar(after_counts.index.astype(str), after_counts.values, color='lightcoral')
ax[1].set_title("After Resampling")
ax[1].set_xlabel("Churn")
ax[1].set_ylabel("Count")

plt.tight_layout()
plt.show()
```
